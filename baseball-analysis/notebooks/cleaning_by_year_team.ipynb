```python
import pandas as pd
import numpy as np
```

```python
batting = pd.read_csv('..\data\\raw\Batting.csv')
```

```python
batting.head()
```

```python
timeIndex = batting[batting['yearID']<1900].index
```

```python
batting.drop(timeIndex,inplace=True)
```

```python
batting.head()
```

```python
batting.drop(columns=['playerID','lgID','GIDP','stint','G'],axis=1,inplace=True)
```

```python
batting.head()
```

```python
batting.drop('IBB',axis=1,inplace=True)
```

```python
byYearTeam = batting.groupby(['yearID','teamID']).sum()
```

```python
byYearTeam.head(100)
```

```python
byYearTeam.to_csv('..\data\\interim\ByYearTeam.csv')
```

```python
# Looking for year in which strikeouts were first counted
nullSearchK = byYearTeam[byYearTeam['SO']>0].index[0]
```

```python
# Post-1910 strikeout counts will be not null
nullSearchK
```

```python
# Same as above, but with caught-stealing
nullSearchCS = byYearTeam[byYearTeam['CS']>0].index[0]
```

```python
# Post-1914 caught-stealing counts will be not null
nullSearchCS
```

```python
# '' Sacrifice flies
nullSearchSF = byYearTeam[byYearTeam['SF']>0].index[0]
```
