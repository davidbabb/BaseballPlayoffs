```python
def playoffs(team):
    for i in range(1900,2016):
        if data[data['teamID']==team][data['yearID']==i]['teamID'].equals(playoff_teams[playoff_teams['yearID']==i]['teamIDwinner']):
            return 1
        elif data[data['teamID']==team][data['yearID']==i]['teamID'].equals(playoff_teams[playoff_teams['yearID']==i]['teamIDloser']):
            return 1
        else:
            return 0
```

```python
# gotta be inside a loop somehow
data['PlayoffTeam'] = data['teamID'].apply(lambda x: playoffs(x))
```
