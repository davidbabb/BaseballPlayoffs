```python
playoff_teams = pd.read_csv('..\data\\raw\SeriesPost.csv')
```

```python
playoff_teams.head()
```

```python
timeIndex = playoff_teams[playoff_teams['yearID']<1900].index
```

```python
playoff_teams.drop(timeIndex,inplace=True)
```

```python
playoff_teams.set_index('yearID',inplace=True)
```

```python
# FUNCTION THAT DISTINGUISHES PLAYOFF TEAMS FROM NON-PLAYOFF TEAMS BY PLACING A 1 IN A NEW COLUMN CREATED IN THE CELL BELOW
# LOCATES EACH INDEX INDIVIDUALLY AND SEARCHES EACH TEAM BY INDEX IN THE PLAYOFF TEAMS DATAFRAME

def playoffs(team,i):
    if i < 1969:
        try:   
            if data[data['teamID']==team]['teamID'].loc[i] == playoff_teams.loc[i]['teamIDwinner']:
                return 1
            elif data[data['teamID']==team]['teamID'].loc[i] == playoff_teams.loc[i]['teamIDloser']:
                return 1
            else:
                return 0
        except:
            return 0
    else:
        try:   
            if data[data['teamID']==team]['teamID'].loc[i] in playoff_teams.loc[i]['teamIDwinner'].values:
                return 1
            elif data[data['teamID']==team]['teamID'].loc[i] in playoff_teams.loc[i]['teamIDloser'].values:
                return 1
            else:
                return 0
        except:
            return 0
```

```python
data['PlayoffTeam'] = 0
```

```python
# gotta be inside a loop somehow
for i in range(1903,2016):
    data['PlayoffTeam'].loc[i] = data['teamID'].loc[i].apply(lambda x: playoffs(x, i))
```
