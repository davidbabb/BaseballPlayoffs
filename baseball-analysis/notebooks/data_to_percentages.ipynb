```python
import pandas as pd
import numpy as np
```

```python
# In this section we will be taking our interim data, converting certain fields to percentages, and adding columns to our dataframe. For example, columns such as 'AB' and 'H' will interact to create 'BA', or batting average, column.
```

```python
masterData = pd.read_csv('..\data\interim\ByYearTeam.csv')
```

```python
masterData.head()
```

```python
# Add batting average column
masterData['BA'] = round(masterData['H']/masterData['AB'],3)
```

```python
masterData.head()
```

```python
# Add plate appearances column
masterData['PA'] = masterData['AB'] + masterData['SF'] + masterData['SH'] + masterData['BB']
```

```python
masterData.head()
```

```python
# Add on-base-percentage column
masterData['OBP'] = round((masterData['H'] + masterData['BB'] + masterData['HBP'])/masterData['PA'],3)
```

```python
masterData.head()
```

```python
# Add slugging percentage column

extraBases = masterData['2B']*2 + masterData['3B']*3 + masterData['HR']*4
singles = masterData['H'] - masterData['2B'] - masterData['3B'] - masterData['HR']
totalBases = singles + extraBases

masterData['SLG'] = round(totalBases/masterData['AB'],3)
```

```python
masterData.head()
```

```python
# Add OPS column
masterData['OPS'] = masterData['OBP'] + masterData['SLG']
```

```python
masterData.head()
```

```python
# Add steal % column

totalSteals = masterData['SB'] + masterData['CS']
masterData['SB%'] = round(masterData['SB']/totalSteals,3)
```

```python
# Deal with 100% steal rates from missing data

def fix(rate):
    if rate == 1:
        return 0
    else:
        return rate
    
masterData['SB%'] = masterData['SB%'].apply(lambda x: fix(x))
```

```python
masterData.head()
```

```python
# Strikeout rate column
masterData['K%'] = round(masterData['SO']/masterData['PA'],3)
```

```python
masterData.head()
```

```python
# Walk rate column
masterData['BB%'] = round(masterData['BB']/masterData['PA'],3)
```

```python
masterData.head()
```

```python
masterData.to_csv('..\data\processed\BattingStats.csv')
```
